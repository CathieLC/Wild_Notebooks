# Mission 5

Rename the columns [of this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv), try using the following 2 different methods to achieve this:
- Setting column names directly when reading the file 
- Editing after creating the dataframe ([This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html))

```
# Setting column names during read_csv import

df_3 = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv", names=['Location', 'Colors', 'Shape', 'State', 'Date'])
df_3.head()
```

```
# Setting column names after read_csv import

df_4 = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv")
df_4.head()
```

```
# Using the DataFrame.rename method, and passing a dict to map the name.
# Note that if a column doesn't need to be renamed, it can be omitted from the dict.

df_4 = df_4.rename(columns={'City': 'Location', 'Colors Reported': 'Colors', 'Shape Reported': 'Shape', 'Time': 'Date'})
df_4.head()
```

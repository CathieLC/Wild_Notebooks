## Mission 4

Create a function using Numpy that solves [second degree equations of this kind](https://www.youtube.com/watch?v=hwq_zpsGixc): $ax^2+bx+c$

```
import numpy as np
```

```
def quadratic_equation_3(a, b, c):
  """
  This function accepts 3 polynomial coefficients a, b, c and solves the equation
  ax² + bx + c = 0 in the form of an array containing the two possible solution.
  """
  coeff_array = np.array([a, b, c])
  return np.roots(coeff_array)

print(quadratic_equation_3(1, -10, 21))

# I manually checked that x² - 10x + 21 = 0 has 3 and 7 as solutions.

print(quadratic_equation_3(1, -3, -10))

# Also checked the solutions manually here
```

## Mission 5

Re-code the Gaussian curve using NumPy functions instead of Python functions.

```
# Code with Python

import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

domain = range(-100,100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))

y = [f(x) for x in domain]
plot = plt.plot(domain, y)
```
